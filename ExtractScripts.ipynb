```python
import pandas as pd
import csv
```

```python
# Read the data, ignoring any errors. It looks like there are only around 20 lines that error
data = pd.read_csv('data/simpsons_script_lines.csv', delimiter=',', quotechar='"', quoting=csv.QUOTE_ALL, error_bad_lines=False)
```

```python
# Get a list of all episode numbers
episodes = set(data.episode_id)
```

```python
# Create scripts for all the episodes
OUTDIR = "episodes_data"
for episode in episodes:
    ep_data = data[data.episode_id == episode]
    ep_data.sort_values('number')
    with open("%s/%04d.txt" % (OUTDIR, episode), 'w') as fd:
        for line in ep_data.raw_text:
            fd.write(line)
            fd.write('\n')
ep_data
```
